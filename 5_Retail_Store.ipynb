```python
retail_data = retail.copy()
```

```python
retail_data = retail_data.drop(columns=['Item_Description','Item_Code'])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
retail_data['YEAR'] = le.fit_transform(retail_data['YEAR'])
retail_data['SUPPLIER'] = le.fit_transform(retail_data['SUPPLIER'])
retail_data['Item_Type'] = le.fit_transform(retail_data['Item_Type'])
```

```python
# Identify numerical columns, including 'YEAR'
numerical_cols = ['Retail_Sales', 'Retail_Transfers', 'Warehouse_Sales']
from sklearn.preprocessing import StandardScaler
# Scale numerical columns
scaler = StandardScaler()
retail_data[numerical_cols] = scaler.fit_transform(retail_data[numerical_cols])
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(retail_data)
# Calculate the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
# Plot the explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), np.cumsum(explained_variance_ratio), marker='o', linestyle='--')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.axhline(y=0.90, color='r', linestyle='-')  # Add a horizontal line at 90%
plt.axhline(y=0.95, color='g', linestyle='-')  # Add a horizontal line at 95%
plt.grid(True)
plt.show()

# The explained variance ratio for each component
for i, ratio in enumerate(explained_variance_ratio):
    print(f"Principal Component {i+1}: {ratio:.4f} (Cumulative: {np.cumsum(explained_variance_ratio)[i]:.4f})")
```

```python
pca = PCA(n_components=5)
retail_pca = pca.fit_transform(retail_data)
retail_pca_df = pd.DataFrame(retail_pca, columns=[f'PC{i+1}' for i in range(5)])
# Print the shape of the transformed dataset to verify the dimensionality reduction
print("Original shape:", retail_data.shape)
print("Transformed shape:", retail_pca.shape)
# Optional: Print the principal components
print("Principal components:\n", pca.components_)
# Optional: Explained variance ratio for the selected components
print("Explained variance ratio for selected components:", pca.explained_variance_ratio_)
# Optional: Cumulative explained variance
print("Cumulative explained variance for selected components:", np.cumsum(pca.explained_variance_ratio_))
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
# Compute WCSS for different number of clusters
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++' )
    kmeans.fit(retail_pca_df)
    wcss.append(kmeans.inertia_)
# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
#model building using k Means
```

```python
retail_pca_df.head(5)
```

```python
#  Apply K-Means clustering
kmeans = KMeans(n_clusters=5, random_state=42)  # Adjust the number of clusters as needed
kmeans_labels = kmeans.fit_predict(retail_pca_df)
```

```python
from sklearn.metrics import silhouette_score
```

```python
from sklearn.utils import resample
# Used a random subset of the data OF 70% 
subset_size = int(0.7 * retail_pca_df.shape[0])
X_subset, clusters_subset = resample(retail_pca_df, kmeans_labels, n_samples=subset_size, random_state=42)
# Calculate silhouette score on the subset
sil_score = silhouette_score(X_subset, clusters_subset)
print(f'Silhouette Score: {sil_score}')
```

```python
# Add the cluster labels to the original dataset
retail['Cluster'] = kmeans_labels
retail.head(5)
```

```python
print("\nCluster Distribution:")
print(retail['Cluster'].value_counts())
```

```python
sns.countplot(data=retail, x='Cluster')
plt.title(" Value Counts of Cluster")
```

```python
# Cluster 1 has highest Count
```

```python
# Sales Analysis by Cluster
cluster_sales = retail.groupby('Cluster').agg({
    'Retail_Sales': ['mean', 'sum'],
    'Retail_Transfers': ['mean', 'sum'],
    'Warehouse_Sales': ['mean', 'sum']
}).reset_index()
# Flatten the MultiIndex columns
cluster_sales.columns = ['Cluster', 
                         'Mean_Retail_Sales', 'Total_Retail_Sales', 
                         'Mean_Retail_Transfers', 'Total_Retail_Transfers', 
                         'Mean_Warehouse_Sales', 'Total_Warehouse_Sales']

print("\nSales Analysis by Cluster:")
print(cluster_sales.to_string(index=False))
```

```python
# Plotting
metrics = ['Mean_Retail_Sales', 'Total_Retail_Sales', 'Mean_Retail_Transfers', 'Total_Retail_Transfers', 'Mean_Warehouse_Sales', 'Total_Warehouse_Sales']

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))
fig.suptitle('Sales Analysis by Cluster', fontsize=16)

for ax, metric in zip(axes.flatten(), metrics):
    ax.bar(cluster_sales['Cluster'], cluster_sales[metric], color='skyblue')
    ax.set_title(metric)
    ax.set_xlabel('Cluster')
    ax.set_ylabel(metric.replace('_', ' '))

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Cluster 0 has highest total sales & mean sales 
```

```python
# Supplier Performance
supplier_sales = retail.groupby('SUPPLIER').agg({
    'Retail_Sales': 'sum',
    'Retail_Transfers': 'sum',
    'Warehouse_Sales': 'sum'
}).reset_index()
supplier_sales['Total_Sales'] = supplier_sales['Retail_Sales'] + supplier_sales['Retail_Transfers'] + supplier_sales['Warehouse_Sales']
supplier_sales['Average_Sales_Per_Item'] = supplier_sales['Total_Sales'] / retail.groupby('SUPPLIER')['Item_Code'].count().values

# Add Cluster column to supplier_sales
supplier_sales['Cluster'] = retail.groupby('SUPPLIER')['Cluster'].first().values

print("\nSupplier Performance:")
print(supplier_sales.to_string(index=False))
```
